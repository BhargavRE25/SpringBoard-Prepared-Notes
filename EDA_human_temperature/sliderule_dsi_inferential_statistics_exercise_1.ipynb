# Is the True Population Average Temperature 98.6$^o$
Bootstrap Performed [Here](#bootstrap) / Frequentist (Central Limit Theorem) Performed [Here](#central limit)

#### Central Limit Theorem Method Analysis
What the Central Limit Theorem allows us to answer is, when a population with parameters: mean $\mu$ = X and standard deviation $\theta$ = Y, is sampled with n samples, what is the probability (the likelihood) that the sample will have mean $\bar{x}$ = Z. Framing this in the context of the current EDA, if the mean temperature of the population is 98.6 degrees what is the likelihood that a sample of n = 130 people from the population will have a mean $\bar{x}$ = 98.25. Since the temperatures are independent of each other, a one sample test is performed. Given that the sample is large with N = 130 and we don't know the standard deviation of the population, we make a best guess assumption that the standard deviation of the sample **s** is representative of, and approximately equal to the standard deviation of the population $\theta$. If we were to sample the population 10,000x, each sample size being N=130, and then computed the mean for each of the 10,000 samples, this distribution would be a normal distribution. According to the Central Limit Theorem, the mean $\bar{x}$ of this distribution of sample means would be equal to the mean $\mu$ of the population. The standard deviation of this distribution $\theta$$\bar{x}$ would be equal to the standard deviation of the population divided by the square root of the sample size $\frac{\theta}{\sqrt{n}}$. Knowing that the distribution of sample means is normal, we can sample the normal distribution using the aforementioned mean $\bar{x}$ and standard deviation $\theta$$\bar{x}$ and then calculate the 95% confidence interval. If our sample's mean of 98.25 is outside this confidence interval, then we ** can reject ** the null hypothesis that 98.6 is the mean population temperature.
> H$_0$: The mean temperature of the population is 98.6$^o$

Process
********
* Sample the normal distribution using mean $\bar{x}$ = 98.6 and standard deviation $\theta$$\bar{x}$ = $\frac{\theta}{\sqrt{n}}$ = $\frac{0.733183}{\sqrt{130}}$ = 0.064304403
* Calculate the 95% confidence interval for this distribution of means
* If our sample mean is outside of this interval then reject the null hypothesis

Using this process we obtain a normal distribution that looks like:
<img src = 'images\clt.png'>
The 95% confidence interval and interpretation for this distribution is:
> **Because our sample mean of 98.25 is outside of the 95% confidence interval (98.475 - 98.723), what Central Limit Theorem tells us is that it is unlikely that 98.6$^o$ is the population mean temperature and that we should reject the null hypothesis. Also, from the qualitative graph above, we can see that the probability of even getting a mean temperature of 98.25$^o$ is extremely close to 0%. This means that if the average population temperature was 98.6, the chance of extracting the sample we have is practically 0% **

#### Boostrap Method Analysis
When performing boostrap analysis of a sample, we treat the sample, as the population and sample it with replacement. In this case, the sample was resampled 10,000x. The mean was calculated for all 10,000 samples and a distribution is plotted with the 95% confidence intervals calculated. If 98.6$^o$ is outside of this interval, then we reject the null hypothesis:
> H$_0$: 98.6$^o$ is the population average temperature

Process
*********
* Resample, with n = 130, the sample 10,000x and create a distribution of 10,000 means
* Calculate the 95% confidence interval for this distribution of means
* If 98.6$^o$ is outside of this interval, then reject the null hypothesis

Using this process we obtain a normal distribtuion that looks like:

<img src = 'images\bootstrap.png'>

The 95% confidence interval and interpretation for this distribution is:
> **Because our hypothesis mean of 98.6$^o$ is outside of the 95% confidence interval (98.121 - 98.375), we should reject the null hypothesis that the population average temperature is 98.6$^o$**

### Is There a Significant Difference in Temperature Between Male and Females
Calculations Performed [Here](#two sample)
**********************
We saw in the initial data inspection that women's temperatures seemed to be on average 0.3$^o$ higher than that of males. If we consider the two samples of 65 males and 65 females as independent samples, we can perform a two sample t-test to see if there truly is a significant difference in average temperatures between males and females, or if this 0.3$^o$ difference was a one-off fluke, only to be seen in this particular sampling. A two sample t-test in essence is performed the same way a one sample test is performed (as we did above). The only difference between the two test is that instead of using the sample mean $\bar{x}$ as the statistic of interest, we use the difference in sample means $\bar{x} = \bar{x}_2 - \bar{x}_1$. And the standard deviation (in this case known as the standard error) is $\theta$ = $s_1$ + $s_2$ = $\sqrt{\frac{\theta_1^2}{n_1} + \frac{\theta_2^2}{n_2}}$. Where these equations come from are out of the scope of this EDA. However the reader can see more information [here](http://onlinestatbook.com/2/sampling_distributions/samplingdist_diff_means.html). The process is the same now as the one sample t-test.
> $H_0$: The mean difference in average temperature between male and female is 0 (meaning males and females have the exact same average temp)

Process
********
* Sample the normal distribution using mean $\bar{x}$ = 0 and standard deviation $\theta$ = $s_1$ + $s_2$ = $\sqrt{\frac{\theta_1^2}{n_1} + \frac{\theta_2^2}{n_2}}$
* Calculate the 95% confidence interval for this distribution of means
* If our two sample mean difference is outside of this interval then reject the null hypothesis

The 95% confidence interval and interpretation for this distribution is:
> **With a calculated 95% confidence interval of -0.245 to  0.248, the sample mean difference of 0.3$^o$ is outside of this interval. The probablity of getting this mean difference was p-value = 0.024 or 2.4% which is below the 5% critical threshold. This means that there is a significant difference in average temperatures between males and females. Females on average do have a higher average temperature than males

```python
import pandas as pd
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats


df = pd.read_csv('data/human_body_temperature.csv')
```

```python
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""

    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n

    return x, y
```

```python
def draw_bs_reps(data, func, size=1):
    """Draw bootstrap replicates."""
    def bootstrap_replicate_1d(data, func):
        return func(np.random.choice(data, size=len(data)))

    # Initialize array of replicates: bs_replicates
    bs_replicates = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_replicates[i] = bootstrap_replicate_1d(data, func)

    return bs_replicates
```

```python
t_test = []
z_test = []

#Perform experiment 5 serpate times
for i in range(5):
    #Take a random sample of 10 temperatures from the total sample
    sample_data_n10 = np.random.choice(temperatures, size=10)
    
    #calculate the mean
    mean = np.mean(sample_data_n10)
    
    #calculate the standard deviation
    std = np.std(sample_data_n10)
    
    #calculate the SE standard mean error (standard deviation of the sampled means)
    mean_sample_std = (std) / np.sqrt(10)
    
    #Perform t-statistics
    t,p = stats.ttest_1samp(sample_data_n10, 98.6)
    #Perform z-statistics
    z_score = (98.6 - mean) / mean_sample_std
    p_value = stats.norm.cdf(98.6, mean, mean_sample_std)
    
    #append results to appropriate list
    t_test.append((t,p))
    z_test.append((z_score , 1-p_value))
    
    
print('t_statistics: ',t_test,'\n\n\nz_statistics: ', z_test)
```

### Two sample test

```python
#Mean of the male sample
male_temp_mean = np.mean(male_temp)
#Mean of the female sample
female_temp_mean = np.mean(female_temp)
#Standard deviation of the male sample
male_temp_std = np.std(male_temp)
#Standard deviation of the female sample
female_temp_std = np.std(female_temp)

#Calculate difference in means
mean_difference = female_temp_mean - male_temp_mean
#Calculate the Standard Error of means
std_difference = np.sqrt((female_temp_std**2 / len(female_temp)) + (male_temp_std**2 / len(male_temp)))

#Sample the normal distribution using the difference in means and the standard error of means
difference_normal = np.random.normal(0, std_difference, 10000)
#Calculate the 95% Confidence interval
percents = np.percentile(difference_normal, [2.5,97.5])
print('95% confidence interval: {}\n'.format(percents))

#calculate the z-score
z_score = (mean_difference - 0) / std_difference
print('z-score: ',z_score)
```

```python
#SciPy two sample t-test
t_test = stats.ttest_ind(female_temp, male_temp,equal_var = True)
print('z-score: {}\np-value: {}'.format(t_test[0], t_test[1]))
```

When utilizing SciPy's two sample test, it yields the same z-score as when the calculation is done manually. Therefore the 95% confidence interval as manually calculated is correct and the mean difference of 0.3 is outside of this interval. this is confirmed by SciPy's p-value of 0.024, meaning there's 2.4% chance of this mean difference happening which is below the 5% critical threshold i.e. outside of the 95% confidence interval
