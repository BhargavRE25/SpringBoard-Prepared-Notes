# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

When trying to model the data with a regression line, we can see that we get a model that can explain 87% of the variance ($R^2$) when utilizing the Predicted values as our independent variable. But with an F-Stastic of almost 8,000, we can conclude that some other indpendent variables would model our data better. Let's run some hypothesis testing. There's clearly some issue with how the "Predicted" values are being predicted. Using boot strap statistics, let's sample the Predicted Readmission Rate values 10,000 times. From this 10,000 values we shall calculate the mean for all 10,000 values. This distribution of sample means will be a normal distribution according to Central Limit Theorem. From there we can calculate the 99% confidence interval of this normal distribution. If the mean of our 'Actual Readmission Rate' is outside of this confidence interval, then we can conclude that data of Predicted Readmission Rates are incorrectly predicted or at least has some error in calculation, and should be further investigated. This is because if the Predicted Values were credible, then sampling them several times would generate some percentage of samples that have a sample mean equivalent of our Actual Readmission Rate mean.

### Bootstrap Analysis and Hypothesis Testing

```python
pred_mean = np.mean(df['Predicted Readmission Rate'])
pred_std = np.std(df['Predicted Readmission Rate'])
standard_error = pred_std / np.sqrt(len(df['Predicted Readmission Rate']))
actual_mean = np.mean(df['Actual Readmission Rate'])
actual_std = np.std(df['Actual Readmission Rate'])

def draw_bs_reps(data, func, size=1):
    """Draw bootstrap replicates."""
    def bootstrap_replicate_1d(data, func):
        return func(np.random.choice(data, size=len(data)))

    # Initialize array of replicates: bs_replicates
    bs_replicates = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_replicates[i] = bootstrap_replicate_1d(data, func)

    return bs_replicates
```

#### Draw 10,000 bootstrap samples from Predicted Readmission Rates and calculate 10,000 means

```python
bs_samples = draw_bs_reps(df['Predicted Readmission Rate'], np.mean, 10000)
percentiles = np.percentile(bs_samples, [0.5,99.5])
fig = plt.figure()
sns.distplot(bs_samples)
plt.text(17.75,6,'99% Confidence Interval: {}'.format(percentiles),size=15)
plt.text(17.75,4,'Mean: {:.2f}'.format(np.mean(bs_samples)),size=15)
plt.text(17.75,5, 'Standard Error: {:.2f}'.format(np.std(bs_samples)), size=15)
plt.text(18.05,6,'Mean of Actual Readmission Rates: {:.2f}'.format(actual_mean),size=15
        )

plt.xlabel('Means of 10,000 Sampled Predicted Readmission Rates')
plt.ylabel('Frequency')
plt.show()
```

# Analysis
------------
> 99% Confidence Interval for Sample Means: 17.85 - 18.12

> Mean of Actual Readmission Rates: 18.23

We can see that our Actual Readmission Rates Mean of 18.23 is outside the 99% CI for sampled means. This means that if the predicted values for Readmission Rates were feasible, then we would have less than a 1% chance of getting the actual Sample of Readmission Rates that we currently have in the data. Because of this, the recommendation would be to investigate further into how the Predicted Readmission Rates are actually predicted. Since they're an integral part of the Excess Ratio, grossly mispredicting a hospitals Readmission Rate would greatly hurt it's predicted vs actual Excess Readmission Ratio
