# Inspect Data

The JSON file has a lot of data in it, and viewing it in its current form is a little tough to read. So first we will clean up the structure of the output so all the levels can be viewed easier

```python
def pretty_print_data(location = 'data/world_bank_projects.json'):
    '''Loads a json file and pretty prints the first observation in the list with an 
    indentation spacing of 4 for better viewing of the data'''
    
    #INPUT - JSON file location
    #OUTPUT - First item in the parsed json list, pretty printed
    
    #Open file: fh
    with open(location) as fh:
        
        #parse json file and save first item in the list as a python object: data
        data = json.load(fh)[0]
        
    #Pretty print first item in the list
    return print(json.dumps(data,indent = 4, sort_keys = True)) 
```

```python
pretty_print_data()
```

# QUESTION 1

### Find the 10 countries with most projects

```python
def question_one(location = 'data/world_bank_projects.json'):
    '''Function parses desired json file and returns a pandas dataframe of the top 10 countries with the 
    most projects and a bar graph showing graphical representation of such'''
    
    #INPUT - JSON DATA LOCATION
    #OUPTUT - DATAFRAME AND BAR GRAPH OF TOP 10 COUNTRIES WITH THE MOST PROJECTS AND 
    
    #parse JSON data into a pandas DataFrame
    data = pd.read_json(location)
    
    # .loc[] - slice dataframe using all rows and only columns 'countryname' and 'countryshortname
    # .groupby().count() - group dataframe by countryshortname and count how many instances of each country there are
    # .sort_values - sort data in descending order
    data = data.loc[:,['countryname','countryshortname']].groupby('countryshortname').count().sort_values('countryname',ascending = False)
    
    
    #return dataframe and bar graph of the top 10 countries with the most projects
    return data[0:10],  data[0:10].plot(kind = 'bar',rot = 45, figsize=(13,8), legend = False)
```

```python
#call function above and answer question
data, ax = question_one()

################# GRAPH FORMATTING###############################
#Change 'Yemen, Repbulic of' to just 'Yemen'
labels = [label.get_text() for label in ax.get_xticklabels()]
labels[4] = 'Yemen'

#Label the graph
ax.set_xticklabels(labels)
ax.set_title('Top 10 Countries with the Most Projects')
ax.set_xlabel('Countries')
ax.set_ylabel('Number of Projects')

#Remove grid and Y-axis labels (chart junk)
ax.grid(False)
ax.get_yaxis().set_ticks([])

#Label each bar with its corresponding value
rects = ax.patches
labels = list(data.countryname)

for rect, label in zip(rects,labels):
    height = rect.get_height()
    ax.text(x = rect.get_x() + rect.get_width()/2, 
            y = height - 1 , 
            s = label,
            ha = 'center', 
            va = 'bottom',
            color = 'white')

################################################################


plt.show()
```

# QUESTION 2

### Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
def question_two(location = 'data/world_bank_projects.json'):
    '''Function takes in JSON data and runs the aggregate function "count" to output the top 
    10 project themes in the JSON data'''
    
    #INPUT -> JSON DATA LOCATION
    #OUTPUT -> DATAFRAME AND BAR GRAPH WITH THE TOP 10 PROJECT THEMES
    
    #Open the file and alias it as object fh
    with open(location) as fh:
        
        #parse the JSON file into a Python object
        data = json.load(fh)
        #parse the data into a pandas DataFrame
        df = json_normalize(data,"mjtheme_namecode")
        
        # .groupby().count() -> group DataFrame by project names and count how many instances there are of each
        # .sort_values() -> sort the count values in descending order
        df = df.groupby('name').count().sort_values('code',ascending = False)
        
    #return the datframe and bar graph of the top 10 project themes
    return df[0:10] , df[0:10].plot(kind = 'bar',rot=90,figsize = (13,8),legend = False)
```

```python
#call the above function and answer the question
data2, ax2 = question_two()


################# GRAPH FORMATTING #####################
ax2.set_title('Top 10 Project Themes')
ax2.set_xlabel('Project Themes')
ax2.set_ylabel('Count')


#Remove grid and Y-axis labels (chart junk)
ax2.grid(False)
ax2.get_yaxis().set_ticks([])

#Label each bar with its corresponding value
rects = ax2.patches
labels = list(data2.code)

for rect, label in zip(rects,labels):
    height = rect.get_height()
    ax2.text(x = rect.get_x() + rect.get_width()/2, 
            y = height - 10 , 
            s = label,
            ha = 'center', 
            va = 'bottom',
            color = 'white')

################################################################

plt.show()
```

# QUESTION 3

### In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
def question_three(location = 'data/world_bank_projects.json'):
    '''Function takes in JSON data and fills in a name for the missing value under project description'''
    
    #INPUT -> JSON DATA LOCATION
    #OUTPUT -> DATAFRAME WITH MISSING PROJECT THEME DESCRIPTION FILLED IN AS: 'Missing Project Theme Description'
    
    #Open the file and alias it as object fh
    with open(location) as fh:
        
        #parse the JSON file into a Python object
        data = json.load(fh)
        #parse the data into a pandas DataFrame
        df = json_normalize(data,"mjtheme_namecode")
        #replace the empty values with desired naming convention
        df.name.replace('','***Missing Project Theme Description***', inplace = True)
        
    #return the dataframe    
    return df
```

```python
#Call function above and answer question
question_three()
```
